```python
import folium
from pprint import pprint

a = folium.Map(location=[52.559860, 1.713940], tiles='cartodbpositron', zoom_start=13)

a
```

```python
from routingpy import HereMaps

api1 = HereMaps(app_id="<your app id>",app_code="<your app code>")
#both the app id and app code can be generated for free on the HereMaps developer site
isochrones = api1.isochrones(locations=[1.713940, 52.559860], profile="car", intervals=[900,1800,2700,3600,4500,5400,6300,7200],departure="now")
#intervals are in seconds, locations require lat and log coordinates to be entered in reverse to be projected onto a folium map
for isochrone in isochrones:
    folium.Polygon(locations=[list(reversed(coord)) for coord in isochrone.geometry],fill="ff0000",opacity=0.5).add_to(a)

a
```

```python
b = folium.Map(location=[52.559860, 1.713940], tiles='cartodbpositron', zoom_start=13)
#this creates another blank map which we can use to add our route to using the Google directions API
coordinates = [[1.712940, 52.5559860], [-0.127630, 51.503391]]
```

```python
from routingpy import Google
api2 = Google(api_key="<your google maps api key>")
#again, you'll need to provide your own google maps api key, which you can generate from the google developer site
route = api2.directions(locations=coordinates,profile="driving",avoid=["tolls"],traffic_model="best_guess",departure_time="now")
folium.PolyLine(locations=[list(reversed(coord)) for coord in route.geometry]).add_to(b)

from datetime import datetime

duration = datetime.fromtimestamp(route.duration).strftime("%I hours %M minutes %S seconds")

print("Duration: {}".format(duration))
print("Distance in km: {}".format(route.distance/1000))
b
```
